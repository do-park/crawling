```python
# ver 4

import cv2, sys
import numpy as np
import os


def unsharp_mask(image, kernel_size=(5, 5), sigma=1.0, amount=1.0, threshold=0):
    """Return a sharpened version of the image, using an unsharp mask."""
    blurred = cv2.GaussianBlur(image, kernel_size, sigma)
    sharpened = float(amount + 1) * image - float(amount) * blurred
    sharpened = np.maximum(sharpened, np.zeros(sharpened.shape))
    sharpened = np.minimum(sharpened, 255 * np.ones(sharpened.shape))
    sharpened = sharpened.round().astype(np.uint8)
    if threshold > 0:
        low_contrast_mask = np.absolute(image - blurred) < threshold
        np.copyto(sharpened, image, where=low_contrast_mask)
    return sharpened


in_path = 'C:/Users/multicampus/Desktop/delete_bg/headwear'
out_path = 'C:/Users/multicampus/Desktop/delete_bg/headwear/png4'

for i in range(0, 351):

    image = cv2.imread(str(i) + '.jpg', cv2.IMREAD_UNCHANGED)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2BGRA)
    image_gray = cv2.imread(str(i) + '.jpg', cv2.IMREAD_GRAYSCALE)
    
    sharpen = unsharp_mask(image_gray)
    edged = cv2.Canny(sharpen, 10, 250)
    

    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (7,7))
    closed = cv2.morphologyEx(edged, cv2.MORPH_CLOSE, kernel)

    contours, _ = cv2.findContours(closed.copy(),cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    total = 0

    # 가장 큰 영역 찾기
    max = 0
    maxcnt = None
    for cnt in contours :
        area = cv2.contourArea(cnt)
        if(max < area) :
            max = area
            maxcnt = cnt

    mask = np.zeros(image.shape, image.dtype)

    # 경계선 내부 255로 채우기
#     cv2.fillPoly(mask, [maxcnt], (255,)*image.shape[2], )
    cv2.fillPoly(mask, contours, (255,)*image.shape[2], )


    masked_image = cv2.bitwise_and(image, mask)

    new_img = cv2.bitwise_and(image, mask)

#     cv2.imwrite("n" + str(i) + '.png', new_img)
    cv2.imwrite(os.path.join(out_path , str(i) + '.png'), new_img)
```
